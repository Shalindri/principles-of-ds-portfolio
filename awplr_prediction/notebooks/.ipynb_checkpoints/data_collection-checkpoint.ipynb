```python
import requests
from bs4 import BeautifulSoup

url = "https://www.cbsl.lk/eResearch/Modules/RD/SearchPages/CMB_LendingAndDeposit.aspx"

# Send a request to the page
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find the table
table = soup.find("table", {"id": "statTB"})

if table:
    print("✅ Table found!")
    print(table.prettify())  # Print table structure
else:
    print("❌ Table NOT found in raw HTML.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from bs4 import BeautifulSoup
import time

# URL
url = "https://www.cbsl.lk/eResearch/Modules/RD/SearchPages/CMB_LendingAndDeposit.aspx"

# Selenium Setup
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run headless
driver = webdriver.Chrome(options=options)

try:
    driver.get(url)
    wait = WebDriverWait(driver, 20)

    # Check if page has iframes
    iframes = driver.find_elements(By.TAG_NAME, "iframe")
    print(f"Number of iframes found: {len(iframes)}")
    if len(iframes) > 0:
        driver.switch_to.frame(iframes[0])  # Switch to the first iframe

    # Set Start Date using JavaScript
    start_date = driver.find_element(By.ID, "ContentPlaceHolder1_txtFrom")
    driver.execute_script("arguments[0].value = '01/01/2008';", start_date)  # MM/DD/YYYY format

    # Set End Date using JavaScript
    end_date = driver.find_element(By.ID, "ContentPlaceHolder1_txtTo")
    driver.execute_script("arguments[0].value = '02/28/2025';", end_date)  # MM/DD/YYYY format

    # Click outside the field to trigger update
    driver.find_element(By.TAG_NAME, "body").click()

    # Wait a bit before clicking "Show"
    time.sleep(2)


    # Click "Show" Button (Using JavaScript Click)
    show_button = driver.find_element(By.ID, "ContentPlaceHolder1_btnShow")
    driver.execute_script("arguments[0].click();", show_button)

    # Wait for the page to update
    time.sleep(120)  # Allow data to load
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # Scroll down

    # Debug: Check Page Source
    print(driver.page_source)  # Look for "statTB"

    # Locate the table
    try:
        table = wait.until(EC.presence_of_element_located((By.ID, "statTB")))
    except:
        print("❌ Table not found within the timeout!")

    # Parse with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")
    table = soup.find("table", {"id": "statTB"})

    if table is None:
        print("❌ Table not found after page load!")
    else:
        print("✅ Table found!")

    # Extract Data
    rows = table.find_all("tr")[2:]  # Skip headers
    data = [[col.text.strip() for col in row.find_all("td")] for row in rows]

    # Convert to DataFrame
    df = pd.DataFrame(data, columns=[
        "End Week", "AWLR", "AWPR (Weekly)", "AWPR (Monthly)", 
        "AWPR (6 months)", "AWDR (Monthly)", "AWDR (6 months)", "AWFDR"
    ])

    # Save to CSV
    df.to_csv("cbsl_lending_deposit_rates.csv", index=False)
    print("✅ Data saved successfully!")

finally:
    driver.quit()
```
